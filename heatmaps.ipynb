```python
from postgre import postgre
import gmaps
```

```python
#we create an instance for our database, in my case my postgre it is running in my localhost i don´t need to
#insert credentials.
dbobject=postgre('','','')
```

```python
#we execute the query statement and we get a DataFrame as a result.
df=dbobject.postgre_to_dataframe('select * from tripadvisor')
df.head(200)
```

```python
#we create a list of tuples with the latitude and longitude as a pair of values.
latitude=df['latitude'].tolist()
longitude=df['longitude'].tolist()
coordinates=[(x,y) for x,y in zip(latitude,longitude) if str(x)!='nan' and str(y)!='nan']
coordinates[:10]
```

```python
#we create our first simple heatmap.
heatmap_layer = gmaps.heatmap_layer(coordinates)
heatmap_layer.point_radius = 15
gmaps.configure(api_key="insert your api key for google maps.")
fig = gmaps.figure()
fig.add_layer(heatmap_layer)
fig
```

```python
#DBSCAN CLUSTERING.
#VARIABLES PRICES AND RATING.
#We create a numerical column whit the rating
def rating(x):
    '''We are going to conver the stars str column into a number'''
    x=str(x)
    if '1' in x:
        return 1
    elif '2' in x:
        return 2
    elif '3' in x:
        return 3
    elif '4' in x:
        return 4
    elif '5' in x:
         return 5 
    else:
        return None
df['rating']=df['stars'].map(lambda x:rating(x))
df['rating'].head(10)
```

```python
##we create a dataframe with the columns filtering all the nan values,and we reset the index after that.
df=df.drop(['address','zipcode','city','services','rooms','min_price','phone_number','created_at_insert','stars']
           ,axis=1)
df.dropna(how='any',inplace=True) 
import numpy as np
df.index=np.arange(len(df))
df.head(10)
```

```python
#we process the algorithim and we create a dataframe with the results.
from sklearn.cluster import DBSCAN
from pandas import DataFrame,merge
X=[["{0:.2f}".format(x1),"{0:.2f}".format(x2)] 
   for x1,x2 in zip(df['rating'],df['max_price'])]
dbscan = DBSCAN()
dbscan = DBSCAN(min_samples=20, eps=3.0)
y_pred = dbscan.fit_predict(X)
cluster=DataFrame(list(y_pred),columns=['cluster_number'])
#we print the lenght ot the list of the results it should be 433, the same as the original dataframe lenght.
len(cluster)
```

```python
#we merge the cluster column and the coordinates column to the original DataFrame.
df=merge(df,cluster,left_index=True, right_index=True)
df.head(10)
```

```python
#we create a layer for each cluster.
print("Los cluster son {0}".format(str(df['cluster_number'].unique())))
#cluster1.
cluster1=df[df['cluster_number']==0]
coordinates=[(x,y) for x,y in zip(cluster1['latitude'].tolist(),cluster1['longitude'].tolist())] 
cluster1_layer=gmaps.symbol_layer(coordinates,fill_color="blue"
                                  ,stroke_color="green",scale=3)
#cluster2.
cluster2=df[df['cluster_number']==1]
coordinates=[(x,y) for x,y in zip(cluster2['latitude'].tolist(),cluster2['longitude'].tolist())]
cluster2_layer=gmaps.symbol_layer(coordinates,fill_color="green",
                                  stroke_color="blue",scale=3)
##cluster3.
cluster3=df[df['cluster_number']==2]
coordinates=[(x,y) for x,y in zip(cluster3['latitude'].tolist(),cluster3['longitude'].tolist())]
cluster3_layer=gmaps.symbol_layer(coordinates,fill_color="red",
                                  stroke_color="red",scale=3)
#nocluster
nocluster=df[df['cluster_number']==-1]
coordinates=[(x,y) for x,y in zip(nocluster['latitude'].tolist(),nocluster['longitude'].tolist())]
nocluster_layer=gmaps.symbol_layer(coordinates,fill_color="white",stroke_color="white",scale=2)
#we create the maps
fig=gmaps.figure()
fig.add_layer(cluster1_layer)
fig.add_layer(cluster2_layer)
fig.add_layer(cluster3_layer)
fig.add_layer(nocluster_layer)
fig
```
